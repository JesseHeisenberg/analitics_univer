# Основные логические операции в Python
В этом ноутбуке рассмотрены основные логические операции, которые используются в Python для сравнения и выполнения условий.

## 1. Логические операторы
Логические операторы используются для выполнения операций над булевыми значениями (`True` или `False`).

### 1.1 Оператор `and` (И)
- Возвращает `True`, если оба выражения истинны.
- Возвращает `False`, если хотя бы одно из выражений ложно.

```
# Пример оператора `and`:
a = True
b = False

print(a and b)  # False
print(a and True)  # True
```

### 1.2 Оператор `or` (ИЛИ)
- Возвращает `True`, если хотя бы одно из выражений истинно.
- Возвращает `False`, если оба выражения ложны.

```
# Пример оператора `or`:
a = True
b = False

print(a or b)  # True
print(False or b)  # False
```

### 1.3 Оператор `not` (НЕ)
- Инвертирует значение: `True` становится `False` и наоборот.

```
# Пример оператора `not`:
a = True

print(not a)  # False
print(not False)  # True
```

## 2. Операторы сравнения
Операторы сравнения используются для сравнения двух значений. Они возвращают `True` или `False`.

### 2.1 Оператор равенства (`==`)
- Проверяет, равны ли два значения.
- Возвращает `True`, если равны, и `False`, если нет.

```
# Пример оператора `==`:
x = 5
y = 10

print(x == y)  # False
print(x == 5)  # True
```

### 2.2 Оператор неравенства (`!=`)
- Проверяет, не равны ли два значения.
- Возвращает `True`, если значения не равны, и `False`, если равны.

```
# Пример оператора `!=`:
x = 5
y = 10

print(x != y)  # True
print(x != 5)  # False
```

### 2.3 Операторы сравнения величин
- `>`: больше
- `<`: меньше
- `>=`: больше или равно
- `<=`: меньше или равно

```
# Примеры:
x = 5
y = 10

print(x > y)   # False
print(x < y)   # True
print(x >= 5)  # True
print(y <= 5)  # False
```

## 3. Комбинирование логических и сравнительных операторов
Логические операторы можно комбинировать с операторами сравнения для создания сложных условий.

```
# Пример:
x = 5
y = 10
z = 15

print(x < y and y < z)  # True
print(x > y or y < z)   # True
print(not (x > z))      # True
```
